### 🔗 Why we need `pyngrok` in Colab?

Colab runs on a remote virtual machine (VM), so any server you start (like FastAPI, Flask, or ADK web UI) binds to `localhost` **within that VM** — not on your local computer. This means you can't access it directly from your browser.

`pyngrok` solves this by creating a **public URL** that tunnels to the internal `localhost`, allowing you to access and share your web app from anywhere.

**Benefits:**
- ✅ Access FastAPI/Flask apps running in Colab
- ✅ Shareable public link for testing or demos
- ✅ No need to set up servers locally

**Note**

- The hospital finder agent runs on Colab’s virtual machine, not your local device.  
- As a result, location-based results (e.g., nearby hospitals) may reflect the server's region (often the USA) rather than your actual location.

### 🔐 How to Get Your `NGROK_AUTHTOKEN`

To use `pyngrok` in Colab, you need a free ngrok authtoken.

#### Steps:
1. Go to [https://dashboard.ngrok.com/signup](https://dashboard.ngrok.com/signup) and create a free account.
2. After logging in, visit [Your Authtoken Page](https://dashboard.ngrok.com/get-started/your-authtoken).
3. Copy the token shown for Command Line.
4. In Colab, Add it as secrets having name as NGROK_AUTHTOKEN and as value paste the ngrok authtoken

**Note**: If you are using in local jupyter notebook then save it as os environment variables.


```python
from google.colab import userdata
import os
os.environ['NGROK_AUTH_TOKEN'] = userdata.get('NGROK_AUTHTOKEN')
```

```python
from pyngrok import ngrok
import uvicorn
import threading

# Expose port 8000
public_url = ngrok.connect(8000)
print("Open this Public URL:", public_url)
```

```python
def run_app():
    uvicorn.run("your_app:app", host="0.0.0.0", port=8000)

thread = threading.Thread(target=run_app)
thread.start()
```
